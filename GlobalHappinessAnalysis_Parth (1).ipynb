```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df15 = pd.read_csv("/content/2015[2].csv")
df16 = pd.read_csv("/content/2016[1].csv")
df17 = pd.read_csv("/content/2017[1].csv")
df18 = pd.read_csv("/content/2018[1].csv")
df19 = pd.read_csv("/content/2019[1].csv")
```

```python
cols = ["Country","Region","Year","Happiness Score","GDP","Family","Health","Freedom","Generosity","Corruption"]

df15 = df15.rename(columns={
    "Economy (GDP per Capita)": "GDP",
    "Health (Life Expectancy)": "Health",
    "Trust (Government Corruption)": "Corruption"
})
df15["Year"] = 2015

df16 = df16.rename(columns={
    "Economy (GDP per Capita)": "GDP",
    "Health (Life Expectancy)": "Health",
    "Trust (Government Corruption)": "Corruption"
})
df16["Year"] = 2016


df17 = df17.rename(columns={
    "Happiness.Score": "Happiness Score",
    "Economy..GDP.per.Capita.": "GDP",
    "Health..Life.Expectancy.": "Health",
    "Generosity_x": "Generosity",
    "Trust..Government.Corruption.": "Corruption"
})
df17["Year"] = 2017
df17["Region"] = None

df18 = df18.rename(columns={
    "Score": "Happiness Score",
    "GDP per capita": "GDP",
    "Social support": "Family",
    "Healthy life expectancy": "Health",
    "Freedom to make life choices": "Freedom",
    "Generosity": "Generosity",
    "Perceptions of corruption": "Corruption",
    "Country or region": "Country"
})
df18["Year"] = 2018
df18["Region"] = None

df19 = df19.rename(columns={
    "Score": "Happiness Score",
    "GDP per capita": "GDP",
    "Social support": "Family",
    "Healthy life expectancy": "Health",
    "Freedom to make life choices": "Freedom",
    "Generosity": "Generosity",
    "Perceptions of corruption": "Corruption",
    "Country or region": "Country"
})
df19["Year"] = 2019
df19["Region"] = None

df15 = df15[cols]
df16 = df16[cols]
df17 = df17[cols]
df18 = df18[cols]
df19 = df19[cols]
```

```python
combined = pd.concat([df15,df16,df17,df18,df19])
```

```python
plt.figure(figsize=(12,8))
region_counts = combined["Region"].value_counts()

plt.pie(region_counts,
        labels=region_counts.index,
        autopct="%1.1f%%",
        startangle=140,
        shadow=True)
plt.title("Region-wise Country Distribution",fontsize=16,fontweight="bold")
```

This chart shows how countries are distributed across different regions in the Happiness dataset.

It helps us understand which region has more countries participating in the survey.

A larger slice means that region contains more countries.

as you can see sub-saharan Africa, Central and Eastern Europe,Latin Amarica and Caribbean Countries more in this data

A Smallest slice means that region contains some contries.

as you can see there Australia and New Zealand Countries some in this data

This information is useful because regional distribution affects the overall happiness comparison.

```python
plt.figure(figsize=(10,6))

year_avg = combined.groupby("Year")["Happiness Score"].mean()

sns.lineplot(x=year_avg.index,y=year_avg.values,color="red",marker="o",markersize=10,linewidth=3)

for i, v in enumerate(year_avg.values):
    plt.text(year_avg.index[i], v + 0.02, f"{v:.2f}",
             ha='center', fontsize=10)

plt.title("Year-wise Average Happiness Score (20152019)",fontsize=16, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Average Happiness Score")

plt.grid(axis='y',alpha=0.4)

sns.despine()
```

This graph shows the average happiness score for each year from 2015 to 2019.

2015 Average Happiness score is 5.36,2016 Average Happiness Score is 5.38, 2017 Average Happiness Score is 5.35, 2018 Average Happiness Score 5.36, 2019 Average Happiness Score is 5.41.

Highest Happiness Average  year is 2019 and lowest Happiness Average Year is 2017

If the line goes up, happiness improved that year.

If the line goes down, average happiness dropped across countries.

```python

year_gdp = combined.groupby("Year")["GDP"].mean()

plt.figure(figsize=(12,8))
sns.lineplot(x=year_gdp.index, y=year_gdp.values, marker="o",markersize=10,color="green" ,linewidth=2)


for i, v in enumerate(year_gdp.values):
    plt.text(year_gdp.index[i], v + 0.02, f"{v:.2f}",
             ha='center', fontsize=10)

plt.title("Year-wise Average GDP", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average GDP", fontsize=12)

plt.grid(axis='both', linestyle='--', alpha=0.4)
sns.despine()
plt.show()
```

This line plot shows how the average GDP of countries in the dataset changed from 2015 to 2019.

A Lowest GDP Average Year is 2015 and Highest GDP Average is 2017.

If the line goes up, GDP improved that year.

```python
plt.figure(figsize=(14,8))
sns.scatterplot(
    data=combined, x="GDP", y="Happiness Score",
    hue="Region",
    palette="tab20",
    s=120,
    edgecolor="black"
)
plt.title("GDP vs Happiness Score (Attractive Scatter Plot)", fontsize=16)
plt.grid(True, alpha=0.5)
```

This scatter plot shows the relationship between a Reginal country's GDP and its happiness score.

We can see that countries with higher GDP usually have higher happiness.

Notice that Which REgion Country Low GDP Which Happiness Score is LOW and Which Country High GDP Which

This indicates that strong economic conditions improve access to better living standards, which increases happiness.

```python

bins = pd.cut(combined['Freedom'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
average_happiness_by_freedom = combined.groupby(bins)['Happiness Score'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.barplot(x='Freedom', y='Happiness Score', data=average_happiness_by_freedom, palette='viridis')
plt.title('Average Happiness Score by Freedom Categories', fontsize=16)
plt.xlabel('Freedom Category', fontsize=12)
plt.ylabel('Average Happiness Score', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```
